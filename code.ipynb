```python
#essential libraries for the project
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#getting the dataset of top 250 anime of 2023 and storing it in a dataset
anime_df = pd.read_csv('top250_anime.csv')
anime_df.head(3)
```

### Lets see what is the distribution of studios in the top 250 anime's and what studios have major holdings in the list
###


```python
studio_counts = anime_df['Studio'].value_counts()

top_n_studio = 10
other_studio_count = studio_counts.iloc[:top_n_studio].sum()

top_studios = studio_counts.iloc[:top_n_studio]
top_studios['Others'] = other_studio_count
```

what we are doing here is trying to see the distribution of studios in top 250 anime in terms of percentage and then plot a pie chart to visualise the data ,to reduce the noise in data we limit it to top 10 studios having most representation in the data while clubing remaining studios as 'others'

### Data Visualisation ###

I make use of basic seaborn and matplotlib to plot the piechart

```python
sns.set_style('whitegrid')
sns.set_palette('Set3')

#axis and figure size
fig, ax = plt.subplots(figsize = (10,10))

ax.pie(top_studios,labels=top_studios.index,autopct='%1.1f%%',startangle=140)

ax.set_title('Top 250 Anime Studio distribution',fontsize='20',fontweight='bold')
ax.set_ylabel('')

# Display the chart with a shadow effect
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax.set_aspect('equal')

ax.legend(title="Studios", loc="upper right")
plt.show()
```

# Let us see the distribution of  genres in top 100 anime 

The code is similar to previous just operated on a subset of top 100 datapoints from 250 and for top genres rather than studios

```python
df = anime_df.head(100)

genre_count = df['Genre'].value_counts()
top_n_genres = 10
other_genre_count = genre_count.iloc[:top_n_genres].sum()

top_genre = genre_count.iloc[:top_n_genres]
top_genre['Others'] = other_genre_count

sns.set_style('whitegrid')
sns.set_palette("Set2")

fig, ax = plt.subplots(figsize =(10,10))

ax.pie(top_genre,labels=top_genre.index,autopct='%1.1f%%',startangle=140)
ax.set_title('Top 100 Anime Genre distribution',fontsize='20',fontweight='bold')

ax.set_ylabel('')
ax.legend(title="Genres:",loc = 'upper right')

plt.show()

#you can play around with the graph by changing the value of .head("") on ln1

```
